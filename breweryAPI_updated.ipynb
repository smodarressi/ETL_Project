```python
# Dependencies
import requests
import pandas as pd
import json
from pprint import pprint
import api_keys
```

```python
#list to iterate through
#recieved from https://www.downloadexcelfiles.com/us_en/download-excel-file-list-cities-california-state
city_list_ca = []
cities_csv = pd.read_csv("list-cities-california-state-us-168j.csv", encoding = "ISO-8859-1")
#name column is an object type, need to convert to string
#get a list from it
cities_ca = cities_csv["Name"].astype(str)
for city_number in range(len(cities_ca)):
    city_list_ca.append(cities_ca[city_number])

print(city_list_ca)
```

```python
#iterate through cities to get all records of breweries in california
#api is limited to 20 max results, a list of cities is needed to iterate through
count = 0
for city in city_list_ca:
    count += 1
    print(f"Getting {count}: California Brewery in {city}")
    print(">>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>")
    url = f"https://api.openbrewerydb.org/breweries?by_city={city}&by_state=California"
    brewery1 = requests.get(url).json()
    print(f"Saving {city} brewery to text file")
    print(">>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>")
    with open('ca_brewery.txt', 'a+') as f:
        f.write("{}\n".format(brewery1))
```

```python
#parsing through the json and reformating the list is difficult
#there were special characters and random indents
#there were also apostrophes that conflicted with the json format
#api isn't received cleanly
#it has to be done manually.....
#txt file was converted to json for import

#file made by iteration through the openbreweryapi was converted to json.
brewery = pd.read_json("ca_brew1.json")
brewery.head()
```

```python
#need to iterate through the list of brewery to get the ids
#first need to call the api and get the business id proprietary to yelp
names_from_json = []
yelp_id_list = []
count = 0

for num in range(len(brewery)):
    count += 1
    brewery_names_json = brewery['name'][num]
    lats = brewery['latitude'][num]
    longs= brewery['longitude'][num]
    print(f"Searching for: {brewery_names_json}")
    print("---------------------------------------")
    try:
        url1 = f'https://api.yelp.com/v3/businesses/search?term={brewery_names_json}&longitude={longs}&latitude={lats}'
        headers = {'Authorization': f"Bearer {api_keys.sm_yelp_api_keys}"}
        yelp_id_search_results = requests.get(url1, headers=headers).json()

        print(f"Found result for {count}: {brewery_names_json}")
        print("---------------------------------------")

        yelp_id_received = yelp_id_search_results['businesses'][0]['id']

        names_from_json.append(brewery_names_json)
        yelp_id_list.append(yelp_id_received)
    except:
        print("No brewery found using latitude, longitude requirement.")
        print("Skipping....")
```

```python
#recall api to make review the listings received
#get the name, ratings, review count and status

yelp_name = []
yelp_b_rating = []
yelp_b_review_count = []
yelp_b_closed_status = []

for num in range(len(yelp_id_list)):
    id_yelp1 = yelp_id_list[num]
    print(f"Finding id {num+1} from list")
    print("------------------------------------")
    try:
        yelp_id_url = f'https://api.yelp.com/v3/businesses/{id_yelp1}'
        headers = {'Authorization': f"Bearer {api_keys.sm_yelp_api_keys}"}
        yelp_id_info = requests.get(yelp_id_url, headers=headers).json()
        name_on_yelp = yelp_id_info["name"]
        print(f"Found {name_on_yelp} associated with id")
        yelp_name.append(yelp_id_info["name"])
        yelp_b_rating.append(yelp_id_info["rating"])
        yelp_b_review_count.append(yelp_id_info["review_count"])
        yelp_b_closed_status.append(yelp_id_info["is_closed"])
    except:
        print("Something bad happened. ABORT!!!!")
```

```python
print(len(names_from_json))
print(len(yelp_id_list))
print(len(yelp_name))
print(len(yelp_b_rating))
print(len(yelp_b_review_count))
print(len(yelp_b_closed_status))
```

```python
#make the dataframe
yelp_brewery_df = pd.DataFrame({
    "names_on_brewery_api":names_from_json,
    "name_on_yelp":yelp_name,
    "yelp_business_id": yelp_id_list,
    "yelp_rating": yelp_b_rating,
    "yelp_review_count": yelp_b_review_count,
    "yelp_closed_status": yelp_b_closed_status   
})
yelp_brewery_df.head()
```
